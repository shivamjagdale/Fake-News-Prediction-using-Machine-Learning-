```python
import pandas as pd
import numpy as np

import re

import nltk

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
nltk.download("stopwords")
```

```python
print(stopwords.words("english"))
```

```python
a = pd.read_csv("/content/drive/MyDrive/ML Datasets/Text data feature extraction .csv")
df = pd.DataFrame(a)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna("")
```

```python
df.isnull().sum()
```

```python
df["content"] = df["author"] + " " + df["title"]
```

```python
df
```

```python
port_stemmer = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub("[^a-zA-Z]", " ", content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stemmer.stem(word) for word in stemmed_content if not word in stopwords.words("english")]
  stemmed_content = " ".join(stemmed_content)
  return stemmed_content
```

```python
df["content"] = df["content"].apply(stemming)
```

```python
df["content"]
```

```python
X = df["content"].values
```

```python
Y = df["label"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y, shuffle=True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(X_train)
```

```python
X_train = vectorizer.transform(X_train)
```

```python
X_test = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction = model.predict(X_train)
X_train_acc = accuracy_score(X_train_prediction, Y_train)
print(X_train_acc)
```

```python
X_test_prediction = model.predict(X_test)
X_test_acc = accuracy_score(X_test_prediction, Y_test)
print(X_test_acc)
```

# **Here we just loaded the saved stuff above and we passed exactly as we did above, stemming first then vectorizer then prediction**

```python
import gradio as gr
import joblib
import string
from nltk.stem.porter import PorterStemmer

# Load saved model and vectorizer
model = joblib.load("news_prediction.pkt")
vectorizer = joblib.load("vectorizer.pkl")
stemmer = joblib.load("stemming.pkt")

# Prediction function
def predict_news(article_text):
    processed_text = stemmer(article_text)
    vectorized = vectorizer.transform([processed_text])
    prediction = model.predict(vectorized)
    return "✅ Real News" if prediction == 0 else "❌ Fake News"

# Gradio interface
interface = gr.Interface(
    fn=predict_news,
    inputs=gr.Textbox(label="Paste News Article Text Here", lines=10, placeholder="Type or paste the full article..."),
    outputs=gr.Text(label="Prediction"),
    title="📰 Fake News Detector",
    description="Paste a news article and find out if it's real or fake. This model uses TF-IDF and Logistic Regression."
)

interface.launch()
```
